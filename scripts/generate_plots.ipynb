## Ablation study expval_budget on GSAT

```python
##
#   Plotting rejection ratios based on expval_budget for different metrics on BAColorGV for GSAT
# The following values were computed by the script 'faithfulness_ablation_expval_budget.sh' and manually reported here
##

budgets     = [10, 20, 50, 80, 100, 150, 200, 300, 500, 1000]
rfidms      = [10, 10, 11, 12, 12, 12, 12, 12, 13, 13]
# rfidps      = []
suff_cause  = [35, 41, 49, 52, 54, 56, 58, 60, 63, 66]
suff        = [1 , 1, 3, 4, 4, 6, 6, 9, 10, 12]
# nec         = []
counter_fid = [50, 49, 50, 50, 50, 50, 50, 50, 50, 50]
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('seaborn-whitegrid')

fig, ax = plt.subplots(figsize=(10, 4))

ax.plot(budgets, rfidms, label='RFID-', color='C0', linestyle='--', linewidth=2)
ax.plot(budgets, counter_fid, label='CF', color='C1', linestyle=':', linewidth=2)
ax.plot(budgets, suff, label='SUF', color='C2', linestyle='-.', linewidth=2)
ax.plot(budgets, [12]*len(budgets), label='FID+', color='C4', linestyle='dotted', linewidth=2)
ax.plot(budgets, [15]*len(budgets), label='FID-', color='C5', linestyle='dotted', linewidth=2)
ax.plot(budgets, suff_cause, label='EST', color='C3', linewidth=2)
# ax.plot(budgets, [0.5] * len(budgets), color='grey', linestyle='-', linewidth=1)

ax.set_xlabel('budget of perturbations', fontsize=14)
ax.set_ylabel('rejection ratio', fontsize=14)
ax.set_ylim((0.0, 100.0))

ax.legend(fontsize=11, loc='best')

ax.tick_params(axis='both', which='major', labelsize=10)
plt.tight_layout()

plt.savefig("../GOOD/kernel/pipelines/plots/expval_budget_ablation_GSAT_BAColorGV.pdf")
plt.show()
```
